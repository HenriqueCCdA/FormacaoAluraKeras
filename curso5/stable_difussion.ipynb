```python
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
modelo = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
```

```python
imagens = modelo.text_to_image('Cat wearing blue jeans', batch_size=3)
```

```python
def plot_images(imagems):
    plt.figure(figsize=(20, 20))
    for i in range(len(imagens)):
        ax = plt.subplot(1, len(imagens), i + 1)
        plt.imshow(imagens[i])
        plt.axis("off")
```

```python
plot_images(imagens)
```

```python
imagens = modelo.text_to_image(
    'Humanoid cat wearing golden jeans, dark fantasys art,'
    'High quality, highly detaild, elegant, sharp focus,'
    'concept art, character concepts, digital painting, mystery, adventure',
    batch_size=3,
)
```

```python
plot_images(imagens)
```
